```python
import numpy as np
np.random.seed(123)

x = np.arange(0, 100, 1)
y = np.random.randint(0, 10, 100) + x #min, max, number of values to generate

plt.plot(x,y)
plt.title("Líneas")
plt.xlabel('Tiempo')
plt.ylabel('Y')
plt.show()
```
